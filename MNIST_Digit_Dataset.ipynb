Training

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
transform = transforms.ToTensor()

train_data = datasets.MNIST('./data', train=True, download=True, transform=transform)
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
```

```python
model = nn.Sequential(
    nn.Conv2d(1, 8, kernel_size=3),
    nn.ReLU(),
    nn.MaxPool2d(2),

    nn.Flatten(),

    nn.Linear(8 * 13 * 13, 128),
    nn.ReLU(),
    nn.Linear(128, 10)
)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
for epoch in range(5):
    for X, Y in train_loader:

        optimizer.zero_grad()      # clear gradients

        output = model(X)          # forward
        loss = loss_fn(output, Y)  # loss

        loss.backward()            # backprop
        optimizer.step()           # update weights

    print("Epoch:", epoch, "Loss:", loss.item())
```

Training Using Class Based model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
transform = transforms.ToTensor()

train_data = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_data = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_data, batch_size=32)
```

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv = nn.Conv2d(1, 8, kernel_size=3)   # (1 → 8 channels)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(2)

        self.fc1 = nn.Linear(8 * 13 * 13, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(self.relu(self.conv(x)))
        x = x.view(x.size(0), -1)   # flatten
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
model = MyModel()

loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
for epoch in range(5):
    for X, Y in train_loader:

        optimizer.zero_grad()       # clear old gradients

        output = model(X)           # forward pass
        loss = loss_fn(output, Y)   # calculate loss

        loss.backward()             # backpropagation
        optimizer.step()            # update weights

    print("Epoch:", epoch, "Loss:", loss.item())
```

Testing using Class Based Model

```python
correct = 0
total = 0

with torch.no_grad():
    for X, Y in test_loader:
        output = model(X)
        _, predicted = torch.max(output, 1)

        total += Y.size(0)
        correct += (predicted == Y).sum().item()

print("Accuracy:", correct / total)
```

```python
with torch.no_grad():
    for X, Y in test_loader:
        output = model(X)
        _, predicted = torch.max(output, 1)

        for i in range(10):
            print(f"Actual: {Y[i].item()}  |  Predicted: {predicted[i].item()}")
        break
```
